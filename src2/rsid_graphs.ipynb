```python
import matplotlib.pyplot as plt
import dist

DIST_PATH = '../data/opensnp/out/rsids_build37_680samples.csv'
distribution = dist.rsid.from_csv(DIST_PATH)

normalized_data = distribution.data / distribution.count
plt.hist(normalized_data.values, bins=50, color='salmon')
plt.xlabel('% of users')
plt.ylabel('# of RSIDs')
plt.title('Distribution of RSIDs across genotype sample (build 37)')
plt.show()
```

```python
filtered_data = normalized_data[normalized_data['count'] > 0.9]
plt.hist(filtered_data.values, bins=50, color='salmon')
plt.xlabel('% of users')
plt.ylabel('# of RSIDs')
plt.title('Distribution of RSIDs across genotype samples (build 37) (>0.9)')
plt.show()
```
