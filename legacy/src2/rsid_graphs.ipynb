```python
import matplotlib.pyplot as plt
import dist

OUT_PATH = '../data/opensnp/out'
distribution = dist.rsid.from_csv(OUT_PATH, build=37)

normalized_data = distribution.data / distribution.count
plt.hist(normalized_data.values, bins=50, color='salmon')
plt.xlabel('% of users')
plt.ylabel('# of RSIDs')
plt.title('Distribution of RSIDs across genotype sample (build 37)')
plt.show()
```

```python
filtered_data = normalized_data[normalized_data > 0.9]
plt.hist(filtered_data.values, bins=50, color='salmon')
plt.xlabel('% of users')
plt.ylabel('# of RSIDs')
plt.title('Distribution of RSIDs across genotype samples (build 37) (>0.9)')
plt.show()
```
